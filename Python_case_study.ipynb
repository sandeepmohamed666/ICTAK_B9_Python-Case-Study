## Question no.1

```python
# Account information stored using variables
account_holder = "John Doe"
account_balance = 1000.0   # Initial balance

# Function to deposit money
def deposit(amount):
    global account_balance
    if amount > 0:
        account_balance += amount
        print(f"Deposited: ₹{amount}")
    else:
        print("Invalid deposit amount!")

# Function to withdraw money
def withdraw(amount):
    global account_balance
    # Check for sufficient balance using conditional statement
    if amount <= 0:
        print("Invalid withdrawal amount!")
    elif amount > account_balance:
        print("Insufficient balance!")
    else:
        account_balance -= amount
        print(f"Withdrawn: ₹{amount}")

# Function to check balance
def check_balance():
    print(f"Current Balance: ₹{account_balance}")

# Menu-driven program
def main():
    while True:
        print("\n--- Bank Account Manager ---")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            amount = float(input("Enter amount to deposit: "))
            deposit(amount)

        elif choice == "2":
            amount = float(input("Enter amount to withdraw: "))
            withdraw(amount)

        elif choice == "3":
            check_balance()

        elif choice == "4":
            print("Thank you for using the bank system!")
            break

        else:
            print("Invalid choice! Please try again.")

# Run the program
main()
```

```python
# Weather Data Analyzer

# a & b. Store weather data (temperature in °C, humidity in %)
temperatures = [32, 24, 20 , 30, 29, 31, 40]
humidity = [60, 85, 75, 70, 80, 75, 40]
rainfall = [0, 5, 10, 0, 20, 0, 5]  # in mm


# d. Function to calculate average
def calculate_average(data):
    return sum(data) / len(data)


# c. Function to determine weather condition
def weather_condition(temp, hum, rain):
    if rain > 15:
        return "Rainy"
    elif temp > 33 and hum < 60:
        return "Sunny"
    elif hum > 75:
        return "Humid"
    else:
        return "Cloudy"


# Calculate averages
avg_temp = calculate_average(temperatures)
avg_humidity = calculate_average(humidity)

# Print averages
print("Average Temperature:", avg_temp, "°C")
print("Average Humidity:", avg_humidity, "%")

print("\nDaily Weather Conditions:")
# Analyze each day's weather
for i in range(len(temperatures)):
    condition = weather_condition(temperatures[i], humidity[i], rainfall[i])
    print(f"Day {i+1}: {condition}")
```
