```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("auto-mpg-1.csv")
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
# Fix for FutureWarning: Assign the result back to the column instead of using inplace=True
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())

# Corrected column name from 'model_year' to 'model year'
summary = df.groupby("model year", as_index=False)[["mpg", "weight", "horsepower"]].mean()

print(summary)

summary.rename(columns={
    "mpg": "avg_mpg",
    "weight": "avg_weight",
    "horsepower": "avg_horsepower"
}, inplace=True)

# Corrected column name in plot
plt.plot(summary["model year"], summary["avg_mpg"])
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.show()
```

## Question 10

```python
import pandas as pd

# Load dataset
df = pd.read_csv("auto-mpg-1.csv")

# Ensure mpg is numeric (in case of issues)
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# Calculate Q1 and Q3
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Filter outliers
outliers_df = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]

# Select required columns (corrected column names)
outliers_df = outliers_df[['car name', 'mpg', 'model year']]

# Display result
print(outliers_df)
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("auto-mpg-1.csv")

# Convert to NumPy array
arr = df.values
# Get column indices
mpg_idx = df.columns.get_loc("mpg")
hp_idx = df.columns.get_loc("horsepower")
wt_idx = df.columns.get_loc("weight")

# Extract columns and convert to float
mpg = pd.to_numeric(arr[:, mpg_idx], errors='coerce')
hp = pd.to_numeric(arr[:, hp_idx], errors='coerce')
wt = pd.to_numeric(arr[:, wt_idx], errors='coerce')
# Get column indices
mpg_idx = df.columns.get_loc("mpg")
hp_idx = df.columns.get_loc("horsepower")
wt_idx = df.columns.get_loc("weight")

# Extract columns and convert to float
mpg = pd.to_numeric(arr[:, mpg_idx], errors='coerce')
hp = pd.to_numeric(arr[:, hp_idx], errors='coerce')
wt = pd.to_numeric(arr[:, wt_idx], errors='coerce')
# Replace NaN with column means
mpg = np.where(np.isnan(mpg), np.nanmean(mpg), mpg)
hp = np.where(np.isnan(hp), np.nanmean(hp), hp)
wt = np.where(np.isnan(wt), np.nanmean(wt), wt)
# Stack columns into a 2D array
matrix = np.column_stack((mpg, hp, wt))

print(matrix.shape)  # (n_rows, 3)
vector = np.array([1, 0.5, -0.2])

# Dot product
result = np.dot(matrix, vector)

print(result)
```

## Question 9

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("auto-mpg-1.csv")

# Convert to NumPy array
arr = df.values
mpg_idx = df.columns.get_loc("mpg")
wt_idx = df.columns.get_loc("weight")

# Convert to numeric arrays
mpg = pd.to_numeric(arr[:, mpg_idx], errors='coerce')
weight = pd.to_numeric(arr[:, wt_idx], errors='coerce')
# Remove rows where either value is NaN
mask = ~np.isnan(mpg) & ~np.isnan(weight)

mpg_clean = mpg[mask]
weight_clean = weight[mask]
correlation = np.corrcoef(mpg_clean, weight_clean)[0, 1]

print("Pearson Correlation (mpg vs weight):", correlation)
```
