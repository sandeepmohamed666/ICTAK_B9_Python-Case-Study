## 01. Plot the trend of average MPG over modelyears (1970-1982) , Which era saw the biggest improvement?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")

# Group by model year and calculate average MPG
avg_mpg = data.groupby("model year")["mpg"].mean()

plt.figure(figsize=(7,4))
plt.plot(avg_mpg.index, avg_mpg.values, marker='o')

plt.title("Average MPG Trend (1970–1982)")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")

plt.grid(True)
plt.show()
```

```python
"""
Inference :-

- Which era saw the biggest improvement?
   The biggest improvement in MPG occurs during the mid-to-late 1970s (around 1974–1978).

- Explanation
   Early 1970s → MPG is relatively low and stable
   Around 1974 onward, MPG starts increasing sharply
   Late 1970s → noticeable improvement in fuel efficiency

- Reason (Real-world context)
   Oil crisis (1973–74) forced manufacturers to build more fuel-efficient cars
   Shift from heavy, large engines → lighter, efficient vehicles

- Final Answer
   Mid-to-late 1970s (especially 1974–1978) saw the biggest improvement in MPG.

"""
```

## 02. Show a bar chart of average horsepower for each country of origin (USA:1, Europe:2,Japan:3)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("auto-mpg.csv")

# Convert 'horsepower' to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows where 'horsepower' is NaN (or consider filling them if appropriate)
df.dropna(subset=['horsepower'], inplace=True)

# Group by origin and calculate average horsepower
avg_hp = df.groupby("origin")["horsepower"].mean()

# Rename index for clarity
avg_hp.index = ["USA", "Europe", "Japan"]

# Plot bar chart
plt.bar(avg_hp.index, avg_hp.values)

plt.xlabel("Country of Origin")
plt.ylabel("Average Horsepower")
plt.title("Average Horsepower by Country")

plt.show()
```

## 03. Draw a histogram of MPG values across all cars. Is the distribution symmetric or skewed?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("auto-mpg.csv")

# Clean data (important)
df["mpg"] = pd.to_numeric(df["mpg"], errors="coerce")
df = df.dropna()

# Histogram
plt.hist(df["mpg"], bins=10, edgecolor="black")
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Count")
plt.show()
```

```python
"""
Inference :-

  The MPG distribution is typically right-skewed (positively skewed).

- Explanation:
  Most cars have low to medium MPG values
  Fewer cars have very high MPG
  This creates a long tail on the right side

- Interpretation:
  Peak is around 15–25 MPG
  A small number of fuel-efficient cars extend the distribution toward higher MPG values.

- Final Answer:
  The distribution is not symmetric — it is right-skewed.

"""
```

## 04. Create a scatter plot of horsepower vs. MPG. What kind of relationship exists?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")

plt.figure(figsize=(6,4))
plt.scatter(data["horsepower"], data["mpg"],
            edgecolors='k', facecolors='none')

plt.title("Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")

plt.show()

```

```python
"""
Inference :-

-The relationship is a negative (inverse) correlation:
  As horsepower increases, MPG decreases
  As horsepower decreases, MPG increases

-The graph:
  Points trend downward from left to right
  This shows a clear inverse relationship

-Interpretation:
  High horsepower cars → more fuel consumption → lower MPG
  Low horsepower cars → better fuel efficiency → higher MPG

"""
```

## 05. Display a box plot of MPG grouped by number of cylinders (3, 4, 5,6,8). Which cylinder count has the most outliers?


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")

# Group MPG values by cylinder count
groups = data.groupby("cylinders")["mpg"].apply(list)

# Order of cylinders
order = [3, 4, 5, 6, 8]
mpg_data = [groups[c] for c in order if c in groups]

plt.figure(figsize=(7,5))
plt.boxplot(mpg_data, labels=order)

plt.title("MPG by Number of Cylinders")
plt.xlabel("Cylinders")
plt.ylabel("MPG")

plt.show()
```

```python
"""
Inference :-

+ Which cylinder count has the most outliers?
  - 4-cylinder cars typically have the most outliers.

+ Why?
  - 4-cylinder cars are the most common in the dataset
  - They include both:
     very fuel-efficient cars (high MPG)
     some lower MPG models
  - This wider variation leads to more outliers

+ Observations
  - 8 cylinders → low MPG, fewer outliers
  - 6 cylinders → moderate spread
  - 4 cylinders → highest variation → more outliers
  - 3 & 5 cylinders → very few data points

+ Final Answer
  - 4-cylinder cars have the most outliers in MPG.

"""
```

## 06. Build a correlation heatmap for all numeric columns,Which pair is most strongly correlated?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")

# Convert horsepower to numeric (in case of missing '?')
data["horsepower"] = pd.to_numeric(data["horsepower"], errors='coerce')

# Select only numeric columns
num_data = data.select_dtypes(include=['number'])

# Correlation matrix
corr = num_data.corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
"""
Inference :-

- Which pair is most strongly correlated?
    The strongest correlation (by absolute value) is usually:
    Weight ↔ Displacement → strong positive correlation (close to +1)

- Other Strong Relationships
    MPG ↔ Weight → strong negative correlation
    MPG ↔ Displacement → strong negative correlation
    Horsepower ↔ Displacement → strong positive correlation

-  Final Answer
    Most strongly correlated pair:
    Weight and Displacement (strong positive correlation)

-  Interpretation
    Bigger engines (high displacement) → heavier cars
    Heavier cars → lower fuel efficiency

"""
```

## 07. Create a pie chart  showing the proportion of cars from each country of origin

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("auto-mpg.csv")

# Count number of cars by origin
car_counts = df["origin"].value_counts().sort_index()

# Rename for clarity
labels = ["USA", "Europe", "Japan"]

# Plot pie chart
plt.pie(car_counts, labels=labels, autopct="%1.1f%%")

plt.title("Proportion of Cars by Country of Origin")

plt.show()
```

## 08. Use a 2x2 subplot grid to show histograms of mpg, horsepower,weight, and acceleration

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("auto-mpg.csv")

plt.figure(figsize=(10,8))

# 1️⃣ MPG
plt.subplot(2, 2, 1)
plt.hist(data["mpg"], edgecolor='black')
plt.title("MPG Distribution")
plt.xlabel("MPG")
plt.ylabel("Frequency")

# 2️⃣ Horsepower
plt.subplot(2, 2, 2)
plt.hist(data["horsepower"], edgecolor='black')
plt.title("Horsepower Distribution")
plt.xlabel("Horsepower")
plt.ylabel("Frequency")

# 3️⃣ Weight
plt.subplot(2, 2, 3)
plt.hist(data["weight"], edgecolor='black')
plt.title("Weight Distribution")
plt.xlabel("Weight")
plt.ylabel("Frequency")

# 4️⃣ Acceleration
plt.subplot(2, 2, 4)
plt.hist(data["acceleration"], edgecolor='black')
plt.title("Acceleration Distribution")
plt.xlabel("Acceleration")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## 10. Multi-Line plot of average weight and fuel efficiency improved over model years?

```python
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("auto-mpg.csv")

# Handle missing values (important for horsepower/mpg datasets)
df = df.dropna()

# Group by model year and calculate averages
avg_data = df.groupby("model year")[["weight", "mpg"]].mean()

# Plot both lines
plt.plot(avg_data.index, avg_data["weight"], label="Average Weight")
plt.plot(avg_data.index, avg_data["mpg"], label="Average MPG")

plt.xlabel("Model Year")
plt.ylabel("Average Values")
plt.title("Trend of Weight and Fuel Efficiency Over Years")

plt.legend()
plt.show()
```
